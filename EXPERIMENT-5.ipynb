Using Python language do the exploratory data analysis of dataset imported in the lab 4.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Load Dataset
from google.colab import files
uploaded = files.upload()  #Uploading the file
```

```python
#LOADING DATASET
import io
df = pd.read_csv(io.BytesIO(uploaded['wine-quality Red.csv']))
```

```python
#DESCRIPTIVE STATISITICS
print("Descriptive Statistics:\n", df.describe())

```

```python
#CHECKING FOR MISSING VALUES
print("\nMissing Values:\n", df.isnull().sum())

```

```python
#CORRELATION MATRIX
correlation_matrix = df.corr(numeric_only=True)
print("\nCorrelation Matrix:\n", correlation_matrix)
```

```python
#DATA VISUALIZATION
sns.pairplot(df.select_dtypes(include=['float64', 'int64']))
plt.show()

```

```python
#CORRELATION HEATMAP
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap")
plt.show()
```

```python
#DISTRIBUTION OF EACH COLUMN
df.hist(figsize=(10, 10), bins=30)
plt.show()
```

```python
#BOXPLOT TO CHECK FOR OUTLIERS
plt.figure(figsize=(12, 6))
sns.boxplot(data=df.select_dtypes(include=['float64', 'int64']), orient='h')
plt.title("Boxplot of Numerical Columns")
plt.show()
```
